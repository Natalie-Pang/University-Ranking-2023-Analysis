### Data Processing (Changing data values)

```python
print('Unique Values of the "University Ranking:')
print(df['University Rank'].unique())
print("\n")
print('Unique Values of the "Overall Score:')
print(df['OverAll Score'].unique())
```

Need to replace the incorrect data format

```python
def convert_data(data):
    avg = None
    if "–" in data:
        data_range = data.split("–")
        avg = (float(data_range[0]) + float(data_range[1])) / 2
    elif "+" in data:
        avg = float(data.replace("+", "").strip())
    else:
        avg = float(data)
    return avg
    
df['University Rank'] = df['University Rank'].apply(convert_data)
df['OverAll Score'] = df['OverAll Score'].apply(convert_data)
```

```python
df['University Rank'] = df['University Rank'].round().astype(int)
```

```python
df['female'] = df['Female:Male Ratio'].str.split(" : ").str[0]
df['male'] = df['Female:Male Ratio'].str.split(" : ").str[1]
df[['male', 'female']]
```

### Correlation between variables

```python
df_corr = df[['OverAll Score', 'Teaching Score', 'Research Score', 'Citations Score', 'Industry Income Score', 'International Outlook Score']].copy()
corr = df_corr.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot = True)
plt.show()

print(corr['OverAll Score'].round(2).sort_values(ascending = False))
```

From the results of both statistics and the heatmap, it is clear that the 'Overall Score' of each university is significantly impacted by these 3 main factors:
- Research Score (0.86)
- Citations  (0.85)
- Teaching Score (0.8)

Compared to the most significant indicators ('Research Score', 'Citations', and 'Teachinng Score'), 'International Outlook Score' and 'Industry Income Score' only has small impact on the Overall Score of the universities. They have correlation scores of 0.64 and 0.42 respectively.

### Countries with the most number of universities

```python
top_20_uni = df['Location'].value_counts().head(20)
plt.figure(figsize=(12, 8))  
sns.barplot(x=top_20_uni, y=top_20_uni.index)
plt.title('Countries with the largest number of universities', loc = 'center')
plt.xlabel('Number of Universities')
plt.ylabel('Country')
```

From the bar chart of countries with the largest number of universities, it is clear thatthe Unitied States has the most number of universities (166), followed by Japan (112) and United Kingdom (89). 

```python
top_5_uni = df.head(5)
plt.figure(figsize=(12, 8))  
positions = range(len(top_5_uni))
bar_width = 0.25

# Create bars for Teaching Score and Research Score
plt.bar(positions, top_5_uni['Research Score'], width=bar_width, label='Research Score')
plt.bar([p + bar_width for p in positions], top_5_uni['Citations Score'], width=bar_width, label='Citations Score')
plt.bar([p + 2 * bar_width for p in positions], top_5_uni['Teaching Score'], width=bar_width, label='Teaching Score')

# Set the position of the x ticks
plt.xticks([p + bar_width for p in positions], top_5_uni['Name of University'], rotation=5)

plt.ylim(80, 100)

# Adding labels and title
plt.xlabel('University')
plt.ylabel('Score')
plt.title('Teaching and Research Scores for Top 5 Universities')
plt.legend()
```
